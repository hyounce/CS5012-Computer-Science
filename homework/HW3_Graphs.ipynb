# **Modeling Data Using a Graph.**

In this example we explore using NetworkX to build graphs based on related data. Below are some examples that illustrate how to build a graph using this package.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)

```

```python
# explicitly set positions
pos = {1: (0, 0), 2: (-1, 0.3), 3: (2, 0.17), 4: (4, 0.255), 5: (5, 0.03)}

options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 5,
    "width": 5,
}
```

```python
nx.draw_networkx(G, pos, **options)
# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

# Now try building a directed Graph

```python
G = nx.DiGraph([(0, 3), (1, 3), (2, 4), (3, 5), (3, 6), (4, 6), (5, 6)])

# group nodes by column
left_nodes = [0, 1, 2]
middle_nodes = [3, 4]
right_nodes = [5, 6]

# set the position according to column (x-coord)
pos = {n: (0, i) for i, n in enumerate(left_nodes)}
pos.update({n: (1, i + 0.5) for i, n in enumerate(middle_nodes)})
pos.update({n: (2, i + 0.5) for i, n in enumerate(right_nodes)})

nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

## Question 1.

Would you use a undirected graph or a directed graph to model this data? Explain.

I would use a directed graph to model this data, because it is a better representation of how following works. For example, Dan follows Jane but Jane does not follow Dan. In a directed graph, this relationship is clear because an arrow will point from Dan to Jane, but not vise-versa. If I were to choose an undirected graph, there would just be an edge between Dan and Jane, which does not clearly represent who follows who. 

## Question 2.

Create this graph using NetworkX and plot the resulting graph.

```python
Twitter_G = nx.DiGraph([("Dan", "Jane"), ("Dan", "Bob"), ("Dan", "Sally"), ("Dan", "Sam"), ("Dan", "Tiana"), ("Dan", "Juan"), ("Dan", "Wei"),
                ("Wei", "Dan"), ("Wei", "Bob"), ("Wei", "Sam"),
                ("Bob", "Jane"), ("Bob", "Wei"), ("Bob", "Tiana"),
                ("Tiana", "Jane"), ("Tiana", "Juan"),
                ("Sally", "Bob"),
                ("Jane", "Juan"), ("Jane", "Wei"),
                ("Juan", "Sally"), ("Juan", "Jane"),
                ("Sam", "Jane")])

pos = nx.spring_layout(Twitter_G)
nx.draw_networkx(Twitter_G, pos, node_size=2000)

ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

## Question 3.

Assume you wanted to determine which user was the biggest "social influencer". Which graph measure would you compute and why?

I would measure the PageRank of a node, which measures not only the number of followers someone has, but the quality of those followers. So a user that is followed by other influential users will have a higher PageRank. Since having followers is a sign of social capital, that is already a good metric for measuring who is the biggest "social influencer". Adding in the quality of followers provides even more insight. 

## Question 4.

Compute the measure chosen and determine who in this example was the biggest social influencer. See example below which analyzes facebook social network data. You will need to have data file facebook_combined.txt to complete this example.

```python
# Load the facebook_combined.txt data file
f = open("facebook_combined.txt", "r")
```

```python
# Read in graph data from facebook data txt file.
G_fb = nx.read_edgelist("facebook_combined.txt", create_using = nx.Graph(), nodetype=int)
print(G_fb)
```

```python
# Compute measures of centrality and degree for the nodes. These attributes can
# be used to help analyze the graph. Then the graph is plotted and color and size
# of the nodes are used to indicate size and betweeness, respectively. (Note: This
# is a large data set and may take 5+ minutes to run.)

pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
```

# Nodes with the highest betweeness

```python
# Identify the top 5 nodes with highest betweeness measures.
sorted(betCent, key=betCent.get, reverse=True)[:5]
```

## Determining PageRank of Twitter Network

```python
pagerank_scores = nx.pagerank(Twitter_G)
sorted_pagerank = sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True)
print("Biggest social influencer - Twitter:", sorted_pagerank[0][0])
```
